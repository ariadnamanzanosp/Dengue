```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming the file name is 'combined_dataset_dengue_temperature.csv'
csv_path = 'combined_dataset_dengue_temperature.csv'

# Read the CSV file into a DataFrame
final_combined_dataset = pd.read_csv(csv_path)

# Reshape the dataframe to long format
df_long = final_combined_dataset.melt(id_vars=['pre_2021', 'tem_me_2021', 'pre_2022', 'tem_me_2022', 'pre_2023', 'tem_me_2023'],
                                      value_vars=['DENGUE_2021', 'DENGUE_2022', 'DENGUE_2023'],
                                      var_name='Year', value_name='Dengue_number')

# Adjust the year column for better readability
df_long['Year'] = df_long['Year'].str.replace('DENGUE_', '')

# Add corresponding Precipitation and Temperature columns based on the Year
df_long['Precipitation'] = df_long.apply(lambda row: row['pre_' + row['Year']], axis=1)
df_long['Temperature'] = df_long.apply(lambda row: row['tem_me_' + row['Year']], axis=1)

# Drop the original pre_* and tem_me_* columns as they are no longer needed
df_long = df_long.drop(columns=['pre_2021', 'tem_me_2021', 'pre_2022', 'tem_me_2022', 'pre_2023', 'tem_me_2023'])

# Rename the columns
df_long = df_long.rename(columns={
    'Precipitation': 'Precipitación (mm)',
    'Temperature': 'Temperatura Media (°C)',
    'Dengue_number': 'Número de Casos de Dengue'
})

# Create the pairplot with a built-in palette
sns.pairplot(df_long, vars=['Precipitación (mm)', 'Temperatura Media (°C)', 'Número de Casos de Dengue'], hue='Year', palette="pastel")

# Add a title
plt.suptitle('Dengue, Precipitación (mm), y Temperatura Media (Nivel Nacional 2021-2023)', y=1.02)

# Show the plot

plt.savefig('dengue_precipitacion_temperatura_media_2021-2023.png')
plt.show()
files.download('dengue_precipitacion_temperatura_media_2021-2023.png')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress


# Melt the dataframe to long format
df_long = final_combined_dataset.melt(id_vars=['pre_2021', 'tem_me_2021', 'pre_2022', 'tem_me_2022', 'pre_2023', 'tem_me_2023'],
                                      value_vars=['DENGUE_2021', 'DENGUE_2022', 'DENGUE_2023'],
                                      var_name='Year', value_name='Dengue_number')

# Adjust the year column for better readability
df_long['Year'] = df_long['Year'].str.replace('DENGUE_', '')

# Add corresponding Temperature columns based on the Year
df_long['Temperature'] = df_long.apply(lambda row: row['tem_me_' + row['Year']], axis=1)

# Filter data for each year
data_2021 = df_long[df_long['Year'] == '2021']
data_2022 = df_long[df_long['Year'] == '2022']
data_2023 = df_long[df_long['Year'] == '2023']

# Perform linear regression for each year
reg_2021 = linregress(data_2021['Temperature'], data_2021['Dengue_number'])
reg_2022 = linregress(data_2022['Temperature'], data_2022['Dengue_number'])
reg_2023 = linregress(data_2023['Temperature'], data_2023['Dengue_number'])

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Plot for 2021
sns.regplot(x='Temperature', y='Dengue_number', data=data_2021, ax=axs[0])
axs[0].set_title('2021')
axs[0].text(0.5, 0.9, f'y={reg_2021.slope:.2f}x+{reg_2021.intercept:.2f}', transform=axs[0].transAxes, ha='center')

# Plot for 2022
sns.regplot(x='Temperature', y='Dengue_number', data=data_2022, ax=axs[1])
axs[1].set_title('2022')
axs[1].text(0.5, 0.9, f'y={reg_2022.slope:.2f}x+{reg_2022.intercept:.2f}', transform=axs[1].transAxes, ha='center')

# Plot for 2023
sns.regplot(x='Temperature', y='Dengue_number', data=data_2023, ax=axs[2])
axs[2].set_title('2023')
axs[2].text(0.5, 0.9, f'y={reg_2023.slope:.2f}x+{reg_2023.intercept:.2f}', transform=axs[2].transAxes, ha='center')

plt.tight_layout()
plt.show()
```
