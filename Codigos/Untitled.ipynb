```python
import pandas as pd

# Load the data
file_path = '/Users/ariadnamanzanosperez/Downloads/Proyecto_2/Dengue/dengue_all_data_date_name.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
print(df.head())
```

```python
# Get unique 'ENTIDAD_ASIG' values
entidades = df['ENTIDAD_ASIG'].unique()

# Create a dictionary to hold the results for each 'ENTIDAD_ASIG'
results = {}

# Loop through each 'ENTIDAD_ASIG' and perform the analysis
for entidad in entidades:
    # Filter the dataframe for the current 'ENTIDAD_ASIG'
    df_entidad = df[df['ENTIDAD_ASIG'] == entidad]
    
    # Example analysis: Count of cases
    count_cases = df_entidad.shape[0]
    
    # Store the result in the dictionary
    results[entidad] = count_cases

# Convert the results dictionary to a dataframe for easy viewing
results_df = pd.DataFrame(list(results.items()), columns=['ENTIDAD_ASIG', 'Count_Cases'])

# Display the results
print(results_df)
```
